```python
def Orlicz_function(u):
    res = np.zeros(len(u))
    for i in range(len(u)):
        n = -1
        while True:
            if u[i] > n and u[i] <= n + 1:  # below two congjugated functions?
                res[i] = (n + 1) * u[i] - (n + 1) * n / 2
                # res[i] = n * u[i] - (n) * (n + 1) / 2
                break
            n = n + 1
    return res
```

```python
osm.plot_Phi_p_plus_Psi(Orlicz_function, u_max=10, du=0.01, max_u_on_plots=10)
```
